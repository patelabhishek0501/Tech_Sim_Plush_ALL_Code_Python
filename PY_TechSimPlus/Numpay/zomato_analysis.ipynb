```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("../datasets/zomato.csv")
data.shape
```

```python
data.keys()
```

```python
data.isnull().sum()
```

```python
data["location"].ffill(inplace=True)
```

```python
data.isnull().sum()
```

```python
data['rate'].unique()
```

```python
data['rate'] = data['rate'].replace('NEW',np.nan)
data['rate'] = data['rate'].replace('-',np.nan)
```

```python
data.isnull().sum()
```

```python
data.dropna(how="any",subset="rate",inplace=True)
```

```python
data['rate'].unique()
```

```python
data.isnull().sum()
```

```python
data.shape
```

```python
rating = '4.1/5'
```

```python
float(rating[:3])
```

```python
def formatRateColumn(rate):
    return float(rate[:3])

data['rate'] = data['rate'].apply(formatRateColumn)
```

```python
data.sort_values(by="rate",ascending=False,inplace=True)
```
