```python
age_mean = nba_data['Age'].mean()
age_mean
```

```python
nba_data.fillna({"Age":age_mean },inplace=True)
nba_data
```

```python
nba_data['Fare']
```

```python
nba_data.fillna({'Fare':np.nan},inplace=True)
```

```python
nba_data['Fare'].ffill(inplace=True)
```

```python
nba_data.drop('Cabin',axis=1,inplace=True)
nba_data
```

```python
nba_data.dropna(how="all",inplace=True)
nba_data.dropna(how="any",inplace=True)
nba_data
```

```python
group_Data=nba_data.groupby(['Pclass','Gender'])
group_Data
```
