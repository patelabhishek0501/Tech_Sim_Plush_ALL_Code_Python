```python
price = "3,000.90".split(",")
float("".join(price))
```

```python
def formate_price(price):
    price = price.split(",")
    price_in_float = float("".join(price))
    return price_in_float
```

```python
data["close "]=data["close "].apply(formate_price)
```

```python
rng = pd.date_range(start = "1/1/2024", end = "8/29/2024", freq="B")
rng
#B = Business days
#D = all days
#W = Weekend
```

```python
daily_dates = pd.date_range(start = "1/1/2024", end = "8/29/2024", freq="D",inclusive="right")
```
