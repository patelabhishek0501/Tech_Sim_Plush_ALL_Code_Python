```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
days = [0,1,2,3,4,5,6]
values = [30,25,45,38,20,32,30]
plt.plot(days,values,ls="-.",lw=3,marker="H",ms=15,color="green",mfc="black",mec="red")
plt.xlabel("Days",color="green",size=20)
plt.ylabel("Temp",color="green",size=20)
plt.title("Days V/s Temp Plot", color="blue",size=20)
plt.axis("off")
plt.show()
```

```python
arr = np.arange(1,51)
y = np.random.randint(20,200,50)
size = np.random.randint(20,200,50)
color = np.random.randn(50)
plt.scatter(arr,y,s = size,c = color)
plt.show()
```

```python
days = [1,2,3,4,5,6,7]
low_temp = np.random.randint(20,30,7)
avg_temp = np.random.randint(25,35,7)
high_temp = np.random.randint(35,45,7)
plt.plot(days,low_temp,label="low temp")
plt.plot(days,avg_temp,label="avg temp")
plt.plot(days,high_temp,label="high temp")
plt.legend()
plt.show()


```

```python
values = np.linspace(-3.14,3.14,100)
sin_values = np.sin(values)
cos_values = np.cos(values)
size = np.random.randint(10,100,100)
colors = np.random.randn(100)
plt.scatter(values,sin_values,label="sin",c=colors,s=size)
plt.plot(values,cos_values,ls="--",label="cos")
plt.legend()
plt.axis("off")
plt.show()
```

```python
years = [2018,2019,2020,2021,2022,2023,2024]
profit = [15000, 20000,10000,30000,25000,18000,14000]
plt.bar(years,profit)
plt.xlabel("Years",size=15,c="blue")
plt.ylabel("Profit",size=15,c="blue")
plt.show()
```

```python
man_ages = [10,20,15,30,75,65,40,45,55,80]
woman_ages = [15,12,18,30,65,55,30,62,58,70]
group = [10,20,30,40,50,60,70,80,90]
plt.hist([man_ages,woman_ages],bins=group,color=["black","red"])
plt.show()

```

```python
programming = ["HTML","JS","Python","C/C++","Java"]
values = [50,70,75,35,30]
e = [0,0,0.1,0,0]
plt.pie(values,labels = programming,radius=1,explode=e,
        autopct="%.0f%%",
        colors=["red","yellow","blue","green","orange"],shadow=True)
plt.show()
```
