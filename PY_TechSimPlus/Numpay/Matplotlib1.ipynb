```python
import matplotlib.pyplot as pit
import numpy as np
```

```python
days = [0,1,2,3,4,6]
values = [23,55,34,45,53,34]
pit.plot(days,values,ls="-.",lw=5,marker="*",ms=15,mfc="pink",mec="red")
pit.xlabel("days")
pit.ylabel("Temp")
pit.title("Days vs Temp plot",color="green", size=10)
pit.axis("on") 
pit.axis("off")
pit.show()
```
